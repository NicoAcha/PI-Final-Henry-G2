```python
import pandas as pd

# Cargar archivo
purchases = pd.read_csv("PurchasesFINAL12312016.csv")
```

```python
#veo tipos de datos de las columnas
purchases.info()
```

```python
#modifico columnas de object a fechas (PODate, ReceivingDate, InvoiceDate, PayDate)
purchases["PODate"] = pd.to_datetime(purchases["PODate"])
purchases["ReceivingDate"] = pd.to_datetime(purchases["ReceivingDate"])
purchases["InvoiceDate"] = pd.to_datetime(purchases["InvoiceDate"])
purchases["PayDate"] = pd.to_datetime(purchases["PayDate"])

purchases.dtypes
```

```python
#Verifico datos vacios en columnas
purchases.isnull().sum().sort_values(ascending=False)
```

```python
#Reemplazo los valores faltantes por desconocido ya que no hay forma de saber el volumen de 
#dichos productos
purchases["Size"] = purchases["Size"].fillna("NaN")

```

```python
#Verifico que no hayan valores duplicados
purchases.duplicated().sum()
#No se encuentra ninguno

```

```python
#Por las dudas elimino espacios en blanco que puedan haber antes o despues de la cadena de texto
cols_str = ["InventoryId", "Description", "Size", "VendorName"]
for col in cols_str:
    purchases[col] = purchases[col].str.strip()

#Chequeo para ver que no hayan espacios en blanco
for col in cols_str:
    print(col, purchases[col].str.match(r"^\s|\s$").sum())
```

```python
#Verifico que la columna Size este toda igual escrita y no hayan duplicados tipo 750ml/750mL
purchases["Size"].value_counts().head(10)
```

```python
purchases[
    (purchases["PurchasePrice"] <= 0) |
    (purchases["Quantity"] <= 0) |
    (purchases["Dollars"] <= 0)
]
```

```python
#elimino el producto con precio de compra igual a 0 ya que no figura en otra tabla
purchases = purchases[purchases["Description"] != "The Macallan Double Cask 12"]
```

```python
purchases.to_csv("PurchasesFINAL12312016_clean.csv", index=False)
```

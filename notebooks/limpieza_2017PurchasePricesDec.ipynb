```python
df["Description"] = df["Description"].fillna("Unknown")
df["Size"] = df["Size"].fillna("Unknown")
```

```python
df["Volume"] = pd.to_numeric(df["Volume"],errors="coerce")
```

```python
df = df[[
    "Brand",
    "Description",
    "Price",
    "Size",
    "Volume",
    "Classification",
    "PurchasePrice",
    "VendorNumber",
    "VendorName"
]]
```

## Conclusiones del proceso de limpieza

Se realizó una verificación de los tipos de datos (dtypes) de cada columna y un análisis de valores nulos para evaluar la calidad del dataset.

Durante el proceso se detectaron 5 valores nulos en la columna Volume, los cuales no pudieron ser inferidos a partir de otras columnas (ya que Size figura como “Unknown” en esos casos). Dado que estos registros representan un porcentaje mínimo del total del dataset, se decidió mantenerlos como NaN para evitar introducir sesgo o valores arbitrarios.

Finalmente, se normalizaron los tipos de datos y se ordenaron las columnas para facilitar la lectura y dejar el archivo listo para etapas posteriores de análisis y carga en BigQuery.

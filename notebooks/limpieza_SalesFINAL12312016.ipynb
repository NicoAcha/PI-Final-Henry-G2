```python
import pandas as pd

# Cargar archivo
sales = pd.read_csv("SalesFINAL12312016.csv")
```

```python
#veo tipos de datos de las columnas
sales.info()
```

```python
#modifico columna de object a fechas
sales["SalesDate"] = pd.to_datetime(sales["SalesDate"])
sales.dtypes
```

```python
#Verifico si hay nulos en las columnas
sales.isnull().sum().sort_values(ascending=False)

#Verifico si hay duplicados en las columnas
sales.duplicated().sum()
```

```python
#Por las dudas elimino espacios en blanco que puedan haber antes o despues de la cadena de texto
cols_str = ["InventoryId", "Description", "Size", "VendorName"]
for col in cols_str:
    sales[col] = sales[col].str.strip()

#Chequeo para ver que no hayan espacios en blanco
for col in ["InventoryId", "Description", "Size", "VendorName"]:
    print(
        col,
        sales[col].str.startswith(" ").sum(),
        sales[col].str.endswith(" ").sum()
    )
```

```python
sales.to_csv("SalesFINAL12312016_clean.csv", index=False)
```

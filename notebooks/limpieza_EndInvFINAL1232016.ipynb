```python
#Ver datos nulos por columna

# 1. Calcular la cantidad de nulos por columna
nulos_cantidad = df_inv.isnull().sum()

# 2. Calcular el porcentaje (para entender la gravedad)
nulos_porcentaje = (df_inv.isnull().mean() * 100).round(2)

# 3. Crear un DataFrame
df_nulos = pd.DataFrame({
    "Cantidad Nulos": nulos_cantidad,
    "Porcentaje (%)": nulos_porcentaje
})

# 4. Filtrar para ver SOLO las columnas que tienen algún nulo (ordenadas de mayor a menor)
df_nulos_filtrado = df_nulos[df_nulos["Cantidad Nulos"] > 0].sort_values("Porcentaje (%)", ascending=False)

print("--- REPORTE DE VALORES FALTANTES ---")
display(df_nulos_filtrado)
```

    2. LIMPIEZA Y TRANSFORMACIÓN DE DATOS

        2.1. CAMBIAR TIPOS DE DATOS

```python
#Convertir tipo de dato de endDate a Datatime
df_inv['endDate'] = pd.to_datetime(df_inv['endDate'])
```

        2.2. REEMPLAZO DE NULOS

```python
# Crear un diccionario con la relacion Tienda -> Ciudad
# Usar solo las filas que si tienen ciudad
mapa_ciudades = df_inv.dropna(subset=['City']).set_index('Store')['City'].to_dict()

# Rellenar los valores nulos en City usando el mapa creado
df_inv['City'] = df_inv['City'].fillna(df_inv['Store'].map(mapa_ciudades))

# Verificar que ya no quedan valores nulos
print("Valores nulos restantes en City:", df_inv['City'].isnull().sum())
```

```python

# 1. Cargar ambos archivos (el Inicial tiene las respuestas que le faltan al Final)

ruta_archivo2 = r"G:/Mi unidad/Soy Henry/Proyecto Final/Proyecto final/Datasets/BegInvFINAL12312016.csv"
df_beg = pd.read_csv(ruta_archivo2) # Asegúrate de tener este archivo en la carpeta

# 2. Crear el "Mapa Maestro" de Ciudades usando AMBOS archivos
# Primero sacamos las ciudades del Inv Inicial
mapa_inicial = df_beg.set_index('Store')['City'].to_dict()
# Luego las del Inv Final (por si hubo tiendas nuevas)
mapa_final = df_inv.dropna(subset=['City']).set_index('Store')['City'].to_dict()

# Fusionamos los mapas (el final tiene prioridad, el inicial rellena huecos)
mapa_maestro = {**mapa_inicial, **mapa_final}

# 3. Rellenar ahora sí
df_inv['City'] = df_inv['City'].fillna(df_inv['Store'].map(mapa_maestro))

# 4. Verificar
nulos_restantes = df_inv['City'].isnull().sum()
print(f"Valores nulos: {nulos_restantes}")

# Si todavía queda alguno (ej. Tienda 46 tampoco estaba en BegInv), avísanos:
if nulos_restantes > 0:
    print("¡Alerta! Las siguientes tiendas no aparecen en ninguno de los dos archivos:")
    print(df_inv[df_inv['City'].isnull()]['Store'].unique())
```

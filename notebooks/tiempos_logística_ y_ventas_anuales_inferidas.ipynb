```python
import pandas as pd


ruta_compras = r"D:\Usuario\Escritorio\HENRY 2025\PF\repositorio PF-facosta73\PI-Final-Henry-G2\data\InvoicePurchases_clean.csv"
df_purchases = pd.read_csv(ruta_compras)

print("‚è≥ Calculando tiempos log√≠sticos...")

# 1. Aseguramos que las fechas sean formato DateTime
df_purchases['PODate'] = pd.to_datetime(df_purchases['PODate'])
df_purchases['ReceivingDate'] = pd.to_datetime(df_purchases['ReceivingDate'])

# 2. Fecha de Recepci√≥n MENOS Fecha de Pedido = D√≠as de demora
df_purchases['Tiempo_Entrega_Dias'] = (df_purchases['ReceivingDate'] - df_purchases['PODate']).dt.days

# 3. Estad√≠sticas generales
promedio_entrega = df_purchases['Tiempo_Entrega_Dias'].mean()
max_entrega = df_purchases['Tiempo_Entrega_Dias'].max()

print("\nüöö AN√ÅLISIS DE TIEMPOS DE ENTREGA (LEAD TIME) üöö")
print("-" * 50)
print(f"‚è±Ô∏è Tiempo promedio de la empresa: {promedio_entrega:.1f} d√≠as")
print(f"üê¢ Peor demora registrada: {max_entrega} d√≠as")

# 4. Buscamos a los culpables: ¬øQu√© proveedores tardan m√°s?
# Agrupamos por proveedor, sacamos el promedio de d√≠as y contamos cu√°ntos pedidos tienen
top_lentos = df_purchases.groupby('VendorName').agg(
    Dias_Promedio=('Tiempo_Entrega_Dias', 'mean'),
    Total_Pedidos=('Tiempo_Entrega_Dias', 'count')
)

# Filtramos solo los proveedores recurrentes (m√°s de 5 pedidos) para que el dato sea s√≥lido
top_lentos = top_lentos[top_lentos['Total_Pedidos'] > 5].sort_values(by='Dias_Promedio', ascending=False).head(5)

print("\nüêå TOP 5 PROVEEDORES M√ÅS LENTOS (Con m√°s de 5 pedidos) üêå")
print("-" * 50)
for proveedor, row in top_lentos.iterrows():
    print(f"üì¶ {proveedor}: {row['Dias_Promedio']:.1f} d√≠as en promedio")
```

### 5. Inferencia de Ventas Totales (F√≥rmula COGS)

```python
import pandas as pd

# 1. Cargamos los archivos
ruta_beginv = r"D:\Usuario\Escritorio\HENRY 2025\PF\repositorio PF-facosta73\PI-Final-Henry-G2\data\BegInvFINAL12312016_clean.csv" 
ruta_compras = r"D:\Usuario\Escritorio\HENRY 2025\PF\repositorio PF-facosta73\PI-Final-Henry-G2\data\InvoicePurchases_clean.csv"
ruta_endinv = r"D:\Usuario\Escritorio\HENRY 2025\PF\repositorio PF-facosta73\PI-Final-Henry-G2\data\EndInvFINAL12312016_clean.csv"

df_beginv = pd.read_csv(ruta_beginv)
df_purchases = pd.read_csv(ruta_compras)
df_endinv = pd.read_csv(ruta_endinv)

print("‚úÖ Los 3 archivos cargados con √©xito en la memoria.")

# 2. Calculamos la plata del inventario inicial (Cantidad x Precio)
if 'CapitalInmovilizado' not in df_beginv.columns:
    df_beginv['CapitalInmovilizado'] = df_beginv['onHand'] * df_beginv['Price']
total_beginv = df_beginv['CapitalInmovilizado'].sum()

# 3. Calculamos la plata del inventario final
if 'CapitalInmovilizado' not in df_endinv.columns:
    df_endinv['CapitalInmovilizado'] = df_endinv['onHand'] * df_endinv['Price']
total_endinv = df_endinv['CapitalInmovilizado'].sum()

# 4. Traemos el total de Compras de todo el a√±o
total_compras_anual = df_purchases['Dollars'].sum() 

# 5. LA F√ìRMULA COGS
ventas_totales_inferidas = total_beginv + total_compras_anual - total_endinv

print("\nüïµÔ∏è‚Äç‚ôÇÔ∏è INFERENCIA DE VENTAS ANUALES (F√≥rmula COGS) üïµÔ∏è‚Äç‚ôÇÔ∏è")
print("-" * 60)
print(f"üì¶ Inventario Inicial (1 Ene):   + ${total_beginv/1e6:,.2f} Millones")
print(f"üõí Compras de todo el a√±o:       + ${total_compras_anual/1e6:,.2f} Millones")
print(f"üõë Inventario Final (31 Dic):    - ${total_endinv/1e6:,.2f} Millones")
print("-" * 60)
print(f"üî• MERCADER√çA VENDIDA (Costo):   = ${ventas_totales_inferidas/1e6:,.2f} Millones")
```

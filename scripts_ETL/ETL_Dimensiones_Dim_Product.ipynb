```python
import pandas as pd

# 1. Cargamos todas las fuentes posibles
df_beg = pd.read_csv('../data/BegInvFINAL12312016_clean.csv')
df_end = pd.read_csv('../data/EndInvFINAL12312016_clean.csv') 
df_prices = pd.read_csv('../data/2017PurchasePricesDec.csv') 

print("üîç Recolectando productos de todas las fuentes...")

# 2. Extraemos (Brand, Description, Size) de cada uno
prod_beg = df_beg[['Brand', 'Description', 'Size']]
prod_end = df_end[['Brand', 'Description', 'Size']]
prod_prices = df_prices[['Brand', 'Description', 'Size']]

# 3. Concatenamos (apilamos) las tres listas
full_catalog = pd.concat([prod_beg, prod_end, prod_prices])

# 4. Ahora s√≠, unimos con la info extra (Volume/Classification) de la tabla de precios
# Usamos drop_duplicates primero en full_catalog para que sea m√°s liviano el merge
full_catalog = full_catalog.drop_duplicates(subset=['Brand'])

df_extra = df_prices[['Brand', 'Volume', 'Classification']].drop_duplicates(subset=['Brand'])

dim_product = pd.merge(full_catalog, df_extra, on='Brand', how='left')

# 5. Formato DER Oficial
dim_product = dim_product.rename(columns={'Brand': 'ProductID'})
dim_product = dim_product[['ProductID', 'Description', 'Size', 'Volume', 'Classification']]

# 6. Limpieza final de seguridad
dim_product = dim_product.drop_duplicates(subset=['ProductID']).reset_index(drop=True)

print(f"‚úÖ Dim_product robusta creada.")
print(f"Total de productos √∫nicos (consolidado): {len(dim_product)}")

dim_product.to_csv('Dim_product.csv', index=False)
dim_product.head()
```
